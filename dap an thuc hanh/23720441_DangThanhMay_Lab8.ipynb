### Bài 1

Nghiên cứu mức thu nhập ($X$) và chi tiêu ($Y$) trong một tháng của một công ty. Khảo sát ngẫu nhiên 7 nhân viên, ta thu được bảng số liệu sau:

| X            | 130     | 150      | 133      | 170      | 170      | 210      | 230      |
| :----------: | :-----: | :------: | :------: | :------: | :------: | :------: | :------: |
| **Y**        | **84**  | **120**  | **108**  | **130**  | **130**  | **150**  | **160**  |


a. Dựa vào bảng dữ liệu trên cho biết có xây dựng được mô hình hồi quy hay không? Nếu có hãy xây dựng mô hình hồi quy để ước tính chi tiêu dựa theo thu nhập

b. Dự đoán mức chi tiêu của nhân viên công ty trên nếu biết thu nhập là 200$

```python
import numpy as np
import statsmodels.api as sm

# Dữ liệu thu nhập (X) và chi tiêu (Y)
X = np.array([130, 150, 133, 170, 170, 210, 230])
Y = np.array([84, 120, 108, 130, 130, 150, 160])

# Kiểm tra tương quan
correlation = np.corrcoef(X, Y)[0, 1]
print(f"Hệ số tương quan Pearson giữa thu nhập và chi tiêu: {correlation}")

# Xây dựng mô hình hồi quy tuyến tính
X = sm.add_constant(X)  # Thêm cột 1s vào X để ước lượng hệ số a và b
model = sm.OLS(Y, X).fit()
print(model.summary())

# Dự đoán mức chi tiêu khi thu nhập là 200
X_new = np.array([[1, 200]])
Y_pred = model.predict(X_new)
print(f"Dự đoán mức chi tiêu khi thu nhập là 200$: {Y_pred[0]}")
```

### Bài 2

Tiến hành nghiên cứu mối quan hệ giữa:

- X: giá bán kẻ của thịt gà (nghìn/kg)
- Y: lượng thịt gà tiêu thụ (kg/tháng)

Dữ liệu thu được gồm 10 tháng như sau:

| X            | 35       | 40       | 45       | 40       | 38       | 40       | 45       | 30       | 42       | 38       |
| :----------: | :------: | :------: | :------: | :------: | :------: | :------: | :------: | :------: | :------: | :------: |
| **Y**        | **2.29** | **2.04** | **0.59** | **2.06** | **2.50** | **2.32** | **1.57** | **3.96** | **1.43** | **2.34** |


a. Dựa vào bảng dữ liệu trên cho biết có xây dựng được mô hình hồi quy hay không? Nếu có hãy xây dựng mô hình hồi quy để ước tính lượng tiêu thụ thịt gà theo giá bán.

b. Dự đoán lượng tiêu thụ khi giá bán lẻ là 41(kg/tháng)

```python
import numpy as np
import statsmodels.api as sm

# Dữ liệu giá bán lẻ (X) và lượng tiêu thụ (Y)
X = np.array([35, 40, 45, 40, 38, 40, 45, 30, 42, 38])
Y = np.array([2.29, 2.04, 0.59, 2.06, 2.50, 2.32, 1.57, 3.96, 1.43, 2.34])

# Kiểm tra hệ số tương quan
correlation = np.corrcoef(X, Y)[0, 1]
print(f"Hệ số tương quan Pearson giữa giá bán và lượng tiêu thụ: {correlation}")

# Xây dựng mô hình hồi quy tuyến tính
X = sm.add_constant(X)  # Thêm cột 1s vào X để ước lượng hệ số a và b
model = sm.OLS(Y, X).fit()
print(model.summary())

# Dự đoán lượng tiêu thụ khi giá bán lẻ là 41
X_new = np.array([[1, 41]])
Y_pred = model.predict(X_new)
print(f"Dự đoán lượng tiêu thụ khi giá bán lẻ là 41 nghìn/kg: {Y_pred[0]}")
```

## Bài 3

Dụa trên file dữ liệu `Advertising.csv` bạn hãy dự đoán doanh số bán hàng (`sales`) dựa trên số tiền chi cho các nền tảng tiếp thị khác nhau như `TV`, `Radio` và `Newspaper`

a. Đọc file dữ liệu cho biết các thông tin cơ bản về dữ liệu: kích thước, dữ liệu có giá trị trống hay không, hiện các giá trị thống kê cơ bản của các thuộc tính.

b. Để dự đoán doanh số bằng phương pháp hồi quy tuyến tính đơn giản (Simple Linear Regression) trước tiên cần lựa chọn một trong các thuộc tính còn lại. Theo bạn thuộc tính nào phù hợp để xây dựng mô hình hồi quy. Giải thích và minh họa bằng 2 cách

c. Xây dụng phương trình hồi quy, trực quan mô hình (đường thẳng) tìm được. Theo bạn để đánh giá mô hình trên cần dùng đại lượng nào?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Phần a: Đọc dữ liệu và hiển thị thông tin cơ bản
data = pd.read_csv('Advertising.csv')

# Hiển thị kích thước dữ liệu
print("Kích thước dữ liệu:", data.shape)

# Kiểm tra các giá trị trống
print("\nGiá trị trống trong dữ liệu:")
print(data.isnull().sum())

# Thống kê cơ bản của các thuộc tính
print("\nCác giá trị thống kê cơ bản:")
print(data.describe())

# Phần b: Chọn thuộc tính để xây dựng mô hình hồi quy
# Tính hệ số tương quan giữa các thuộc tính với sales
correlation = data.corr()
print("\nHệ số tương quan với sales:")
print(correlation['sales'])

# Vẽ biểu đồ phân tán cho các thuộc tính với sales để trực quan
# Check the actual column names in your DataFrame
print(data.columns)
# Adjust the column names in x_vars if necessary.
# For example, if the column name is 'radio' instead of 'Radio', change it to:
sns.pairplot(data, x_vars=['TV', 'radio', 'newspaper'], y_vars='sales', height=5, aspect=0.7)
plt.show()

# Lựa chọn thuộc tính: Dựa trên hệ số tương quan, TV có mối quan hệ mạnh nhất với sales

# Phần c: Xây dựng mô hình hồi quy với thuộc tính TV và trực quan hóa
X = data['TV']
y = data['sales']

# Thêm hệ số chặn vào X
X = sm.add_constant(X)

# Xây dựng mô hình hồi quy tuyến tính đơn giản
model = sm.OLS(y, X).fit()
print(model.summary())

# Trực quan hóa đường hồi quy
plt.figure(figsize=(10, 6))
plt.scatter(data['TV'], data['sales'], color='blue', label='Dữ liệu thực tế')
plt.plot(data['TV'], model.predict(X), color='red', label='Đường hồi quy')
plt.xlabel('Chi tiêu cho TV (nghìn USD)')
plt.ylabel('Doanh số (nghìn USD)')
plt.title('Hồi quy tuyến tính giữa TV và sales')
plt.legend()
plt.show()

# Đánh giá mô hình
r_squared = model.rsquared
print(f"Hệ số xác định (R^2) của mô hình: {r_squared}")
```
