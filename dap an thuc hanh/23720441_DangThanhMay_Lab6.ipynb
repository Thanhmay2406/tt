```python
# Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến POP.
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
POP = np.random.uniform(120, 200, 10000)
```

c. Vẽ histogram của hai câu a, b trong cùng một hình để so sánh. Theo bạn hình dạng của histogram thay đổi nói lên điều gì?

Gợi ý: tạo một danh sách rỗng để chứa các trung bình mẫu tính được, sử dụng hàm append() để thêm giá trị trung bình vào danh sách sau mỗi lần tính.

```python
# a. Thực hiện bài 2 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

def PPMEAN(data, sample_size, num_samples):
  sample_means = []
  for _ in range(num_samples):
    sample = np.random.choice(data, size=sample_size, replace=False)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)
  return sample_means


# Sử dụng hàm để lấy 100 trung bình mẫu với kích thước mẫu là 20
sample_means_20 = PPMEAN(data=POP, sample_size = 20, num_samples=100)

# Vẽ histogram cho 100 trung bình mẫu
plt.figure(figsize=(8, 5))
sns.histplot(sample_means_20, bins=10, color='skyblue', stat='density', edgecolor='black')
plt.title('Histogram của 100 trung bình mẫu (cỡ 20)')
plt.xlabel('Trung bình mẫu (cm)')
plt.ylabel('Mật độ')
plt.grid()
plt.show()
```

```python
#  Thực hiện bài 3 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

# Sử dụng hàm để lấy 100 trung bình mẫu với kích thước mẫu là 20
sample_means_100 = PPMEAN(data=POP, sample_size = 100, num_samples=100)

# Vẽ histogram cho 100 trung bình mẫu
plt.figure(figsize=(8, 5))
sns.histplot(sample_means_100, bins=20, color='skyblue', stat='density', edgecolor='black')
plt.title('Histogram của 100 trung bình mẫu (cỡ 100)')
plt.xlabel('Trung bình mẫu (cm)')
plt.ylabel('Mật độ')
plt.grid()
plt.show()
```

### Bài 5: Khảo sát tỷ lệ mẫu

a. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram tỷ lệ người cao của các mẫu

b. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 500. Vẽ histogram tỷ lệ người cao của các mẫu

```python
# a. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram tỷ lệ người cao của các mẫu
POP = np.random.uniform(120,200,10000)

def KSMAU(data, sample_size, num_samples):
  sample_ratios = []
  for _ in range(num_samples):
    sample = np.random.choice(data, size=sample_size, replace=False)
    count_tall = np.sum(sample >= 180)
    ratio_tall = count_tall / sample_size
    sample_ratios.append(ratio_tall)
  return sample_ratios

# Mô phỏng 1000 lần với kích thước mẫu 100
sample_size_100 = 100
num_samples = 1000
tall_ratios_100 = KSMAU(POP,sample_size_100, num_samples)

# Vẽ histogram tỷ lệ người cao
plt.figure(figsize=(10, 6))
sns.histplot(tall_ratios_100, bins=20, color='skyblue', stat='density', edgecolor='black')
plt.title('Histogram tỷ lệ người cao (kích thước mẫu 100)')
plt.xlabel('Tỷ lệ người cao (%)')
plt.ylabel('Mật độ')
plt.grid()
plt.show()
```

```python
# b. Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 500. Vẽ histogram tỷ lệ người cao của các mẫu
# Mô phỏng 1000 lần với kích thước mẫu 500
sample_size_500 = 500
tall_ratios_500 = KSMAU(POP,sample_size_500, num_samples)

# Vẽ histogram tỷ lệ người cao
plt.figure(figsize=(10, 6))
sns.histplot(tall_ratios_500, bins=20, color='lightcoral', stat='density', edgecolor='black')
plt.title('Histogram tỷ lệ người cao (kích thước mẫu 500)')
plt.xlabel('Tỷ lệ người cao (%)')
plt.ylabel('Mật độ')
plt.grid()
plt.show()
```

### Bài 7:  Bài Tập Ước Lượng

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


```python
import numpy as np
import scipy.stats as stats

# Dữ liệu
weights = np.array([2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8])

# a, Biết  σ=0.3

# Tính toán các thông số
n = len(weights)  # Kích thước mẫu
mean_weight = np.mean(weights)  # Trung bình mẫu
sigma = 0.3  # Độ lệch chuẩn đã biết
alpha = 0.05  # Mức ý nghĩa

# Tính khoảng tin cậy
z_score = stats.norm.ppf(1 - alpha / 2)  # Giá trị z cho 95% khoảng tin cậy
margin_of_error = z_score * (sigma / np.sqrt(n))  # Lỗi biên

confidence_interval = (mean_weight - margin_of_error, mean_weight + margin_of_error)
print("Khoảng tin cậy 95% cho cân nặng trung bình (σ đã biết):", confidence_interval)
```

```python
# b, Không biết  σ
# Tính toán độ lệch chuẩn mẫu
sample_std = np.std(weights, ddof=1)  # Độ lệch chuẩn mẫu

# Tính khoảng tin cậy
t_score = stats.t.ppf(1 - alpha / 2, n - 1)  # Giá trị t cho 95% khoảng tin cậy
margin_of_error_t = t_score * (sample_std / np.sqrt(n))  # Lỗi biên

confidence_interval_t = (mean_weight - margin_of_error_t, mean_weight + margin_of_error_t)
print("Khoảng tin cậy 95% cho cân nặng trung bình (σ không biết):", confidence_interval_t)

```

Sử dụng mẫu là điểm thi của 36 học sinh đầu tiên trong bộ dữ liệu, bạn hãy xây dựng khoảng tin cậy **95%** cho điểm SAT trung bình của trường phổ thông đấy. Cho biết khoảng ước lượng bạn xây dựng có đúng không, dựa trên giá trị đã tính ở câu trên.

```python
import pandas as pd

# Đọc dữ liệu từ tệp CSV
data = pd.read_csv('SAT-lard.csv')

# Xem các dòng đầu tiên của dữ liệu
print(data.head())
```

```python
import numpy as np
import scipy.stats as stats

# Lấy điểm thi SAT
sat_scores = data['SAT Score']

# Tính các thông số
n = len(sat_scores)  # Kích thước mẫu
mean_score = np.mean(sat_scores)  # Trung bình mẫu
std_score = np.std(sat_scores, ddof=1)  # Độ lệch chuẩn mẫu
alpha = 0.01  # Mức ý nghĩa cho khoảng tin cậy 99%

# Tính khoảng tin cậy
t_score = stats.t.ppf(1 - alpha / 2, n - 1)  # Giá trị t cho 99% khoảng tin cậy
margin_of_error = t_score * (std_score / np.sqrt(n))

confidence_interval_99 = (mean_score - margin_of_error, mean_score + margin_of_error)
print("Khoảng tin cậy 99% cho trung bình điểm thi SAT:", confidence_interval_99)
```

```python
alpha = 0.10  # Mức ý nghĩa cho khoảng tin cậy 90%

# Tính khoảng tin cậy
t_score = stats.t.ppf(1 - alpha / 2, n - 1)  # Giá trị t cho 90% khoảng tin cậy
margin_of_error = t_score * (std_score / np.sqrt(n))

confidence_interval_90 = (mean_score - margin_of_error, mean_score + margin_of_error)
print("Khoảng tin cậy 90% cho trung bình điểm thi SAT:", confidence_interval_90)
```

```python
# Tính trung bình điểm SAT
mean_score_full = np.mean(sat_scores)
print("Điểm SAT trung bình của trường phổ thông:", mean_score_full)
```

```python
# Lấy điểm thi của 36 học sinh đầu tiên
sample_first_36 = sat_scores[:36]

# Tính toán khoảng tin cậy
n_sample = len(sample_first_36)
mean_sample = np.mean(sample_first_36)
std_sample = np.std(sample_first_36, ddof=1)
alpha = 0.01  # Mức ý nghĩa cho khoảng tin cậy 99%

# Tính khoảng tin cậy
t_score = stats.t.ppf(1 - alpha / 2, n_sample - 1)
margin_of_error_sample = t_score * (std_sample / np.sqrt(n_sample))

confidence_interval_sample_99 = (mean_sample - margin_of_error_sample, mean_sample + margin_of_error_sample)
print("Khoảng tin cậy 99% cho điểm SAT trung bình (36 học sinh đầu tiên):", confidence_interval_sample_99)
```

```python
# Tính khoảng tin cậy 95%
alpha = 0.05  # Mức ý nghĩa cho khoảng tin cậy 95%

# Tính khoảng tin cậy
t_score = stats.t.ppf(1 - alpha / 2, n_sample - 1)
margin_of_error_sample_95 = t_score * (std_sample / np.sqrt(n_sample))

confidence_interval_sample_95 = (mean_sample - margin_of_error_sample_95, mean_sample + margin_of_error_sample_95)
print("Khoảng tin cậy 95% cho điểm SAT trung bình (36 học sinh đầu tiên):", confidence_interval_sample_95)

# Nhận xét
if confidence_interval_sample_95[0] <= confidence_interval_sample_99[0] and confidence_interval_sample_95[1] >= confidence_interval_sample_99[1]:
    print("Khoảng tin cậy 95% bao gồm khoảng tin cậy 99%.")
else:
    print("Khoảng tin cậy 95% không bao gồm khoảng tin cậy 99%.")
```
