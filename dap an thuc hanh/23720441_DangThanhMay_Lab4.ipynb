## HIỂU DỮ LIỆU

1. Đọc dữ liệu
2. Xoá cột `Unnamed: 0`
3. Cho biết dữ liệu có bao nhiêu dòng, bao nhiêu thuộc tính.
4. Trong các thuộc tính, thuộc tính nào là thuộc tính phân loại, thuộc tính nào là thuộc tính định lượng
5. Có bao nhiêu nam, bao nhiêu nữ. Tỷ lệ bao nhiêu?
6. Có bao nhiêu người hút thuốc. Tỷ lệ bao nhiêu?
7. Ngày nào trong tuần khách hàng hay đến dùng bữa nhất. Tỷ lệ bao nhiêu?
8. Nhà hàng phục vụ bữa ăn nào nhiều nhất (Breakfast, Lunch, Dinner)?
9. Nhóm khách hàng dùng bữa tại nhà hàng nhiều nhất là bao nhiêu người?
10. Trung bình trong một bữa ăn nhà hàng phục vụ mấy khách?

```python
#1. Đọc dữ liệu
import pandas as pd

# Đọc dữ liệu từ file tips.csv
df = pd.read_csv('data/tips.csv')

# Xoá cột Unnamed: 0
df = df.drop(columns=['Unnamed: 0'])
```

```python
# 3. Cho biết dữ liệu có bao nhiêu dòng, bao nhiêu thuộc tính.
print(f'Số dòng: {df.shape[0]}')
print(f'Số cột: {df.shape[1]}')

# Số lượng dòng và thuộc tính
num_rows, num_columns = df.shape
print(f"Số lượng dòng: {num_rows}, Số lượng thuộc tính: {num_columns}")
```

```python
# 4. Trong các thuộc tính, thuộc tính nào là thuộc tính phân loại, thuộc tính nào là thuộc tính định lượng
```

Thuộc tính phân loại: sex, smoker, day, time

Thuộc tính định lượng: total_bill, tip, size

```python
# 5. Có bao nhiêu nam, bao nhiêu nữ. Tỷ lệ bao nhiêu?
# Số lượng nam, nữ và tỷ lệ
gender_counts = df['sex'].value_counts()
num_males = gender_counts['Male']
num_females = gender_counts['Female']
male_ratio = num_males / num_rows * 100
female_ratio = num_females / num_rows * 100

print(f"Số lượng nam: {num_males}, Số lượng nữ: {num_females}")
print(f"Tỷ lệ nam: {male_ratio:.2f}%, Tỷ lệ nữ: {female_ratio:.2f}%")
```

```python
# 6. Số lượng người hút thuốc và tỷ lệ
smoker_count = df['smoker'].value_counts()['Yes']
smoker_ratio = smoker_count / num_rows * 100

print(f"Số lượng người hút thuốc: {smoker_count}")
print(f"Tỷ lệ người hút thuốc: {smoker_ratio:.2f}%")
```

```python
# 7. Ngày nào trong tuần khách hàng hay đến dùng bữa nhất. Tỷ lệ bao nhiêu?
day_counts = df['day'].value_counts()
most_frequent_day = day_counts.idxmax()
most_frequent_day_count = day_counts.max()
most_frequent_day_ratio = most_frequent_day_count / num_rows * 100

print(f"Ngày khách hàng đến dùng bữa nhiều nhất: {most_frequent_day}\nTỷ lệ: {most_frequent_day_ratio:.2f}%")
```

```python
# 8. Nhà hàng phục vụ bữa ăn nào nhiều nhất (Breakfast, Lunch, Dinner)?
# Nhà hàng phục vụ bữa ăn nào nhiều nhất
meal_counts = df['time'].value_counts()
most_frequent_meal = meal_counts.idxmax()

print(f"Bữa ăn nào phục vụ nhiều nhất: {most_frequent_meal}")
```

```python
# 9. Nhóm khách hàng dùng bữa tại nhà hàng nhiều nhất là bao nhiêu người?
# Nhóm khách hàng dùng bữa tại nhà hàng nhiều nhất
group_size_counts = df['size'].value_counts()
most_frequent_group_size = group_size_counts.idxmax()

print(f"Nhóm khách hàng dùng bữa nhiều nhất là: {most_frequent_group_size} người")
```

```python
# 10. Trung bình trong một bữa ăn nhà hàng phục vụ mấy khách?
# Trung bình trong một bữa ăn nhà hàng phục vụ mấy khách
average_size = df['size'].mean()

print(f"Trung bình trong một bữa ăn nhà hàng phục vụ: {average_size:.2f} khách")
```

## YẾU TỐ ẢNH HƯỞNG ĐẾN TIỀN TIP

Trong phần này, chúng ta thử tìm hiểu về tiền tip tại nhà hàng và các yếu tố có thể ảnh hưởng đến nó

1. Tỷ lệ nhận được tiền tip của nhà hàng là bao nhiêu?
2. Tỷ lệ tiền tip trên tổng hóa đơn là bao nhiêu?
3. Giữa nam và nữ ai tip nhiều tiền hơn, ai tip thường xuyên hơn?
4. Giữa người hút thuốc và người không hút thuốc ai tip nhiều tiền hơn, ai tip thường xuyên hơn?
5. Thống kê tổng số tiền tip nhận được theo ngày.

```python
# 1. Tỷ lệ nhận được tiền tip của nhà hàng là bao nhiêu?
# Tính tỷ lệ nhận được tiền tip
tip_received_ratio = (df['tip'].count() / num_rows) * 100
print(f"Tỷ lệ nhận được tiền tip: {tip_received_ratio:.2f}%")
```

```python
# 2. Tỷ lệ tiền tip trên tổng hóa đơn là bao nhiêu?
# Tính tỷ lệ tiền tip trên tổng hóa đơn
df['tip_ratio'] = (df['tip'] / df['total_bill']) * 100
average_tip_ratio = df['tip_ratio'].mean()

print(f"Tỷ lệ tiền tip trung bình trên tổng hóa đơn: {average_tip_ratio:.2f}%")
```

```python
# 3. Giữa nam và nữ ai tip nhiều tiền hơn, ai tip thường xuyên hơn?
# Tính tiền tip trung bình theo giới tính
average_tip_by_gender = df.groupby('sex')['tip'].mean()
tip_frequency_by_gender = df['sex'].value_counts()

print("Tiền tip trung bình theo giới tính:")
print(average_tip_by_gender)
print("Số lần tip theo giới tính:")
print(tip_frequency_by_gender)
```

```python
# 4. Giữa người hút thuốc và người không hút thuốc ai tip nhiều tiền hơn, ai tip thường xuyên hơn?
# Tính tiền tip trung bình theo tình trạng hút thuốc
average_tip_by_smoker = df.groupby('smoker')['tip'].mean()
tip_frequency_by_smoker = df['smoker'].value_counts()

print("Tiền tip trung bình theo tình trạng hút thuốc:")
print(average_tip_by_smoker)
print("Số lần tip theo tình trạng hút thuốc:")
print(tip_frequency_by_smoker)
```

```python
# 5. Thống kê tổng số tiền tip nhận được theo ngày.
# Tính tổng tiền tip theo ngày
total_tip_by_day = df.groupby('day')['tip'].sum()

print("Tổng số tiền tip nhận được theo ngày:")
print(total_tip_by_day)
```

## Với Matplotlib

Bạn hãy trực quan với thư viện matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
# đọc dữ liệu
tips = pd.read_csv('data/tips.csv')
print(tips)
```

```python
# 1. Vẽ đồ thị thể hiện tổng tiền tip theo ngày
total_tip_by_day = tips.groupby('day')['tip'].sum()

plt.figure(figsize=(8, 5))
total_tip_by_day.plot(kind='bar', color='skyblue')
plt.title('Tổng tiền tip theo ngày')
plt.xlabel('Ngày')
plt.ylabel('Tổng tiền tip (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Vẽ đồ thị thể hiện tổng tiền tip theo bữa ăn (time)
total_tip_by_time = tips.groupby('time')['tip'].sum()
plt.figure(figsize=(8, 5))
total_tip_by_time.plot(kind='bar', color='lightcoral')
plt.title('Tổng tiền tip theo bữa ăn')
plt.xlabel('Buổi ăn')
plt.ylabel('Tổng tiền tip (USD)')
plt.show()
```

```python
# 3. Vẽ đồ thị histogram của tổng hoá đơn
plt.figure(figsize=(8,5))
plt.hist(tips['total_bill'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram của tổng hoá đơn')
plt.xlabel('Tổng hoá đơn (USD)')
```

```python
# 4. Vẽ đồ thị boxplot của tổng hoá đơn
plt.figure(figsize=(8,5))
plt.boxplot(tips['total_bill'], vert=False, patch_artist=True)
plt.title('Boxplot của tổng hóa đơn')
plt.xlabel('Tổng hóa đơn (USD)')
plt.show()
```

```python
# 5. Vẽ đồ thị để kiểm tra sự liên quan giữa tổng hoá đơn và tiền tip.
plt.figure(figsize=(8,5))
plt.scatter(tips['total_bill'], tips['tip'], color='blue', alpha=0.5)
plt.title('Sự liên quan giữa tổng hóa đơn và tiền tip')
plt.xlabel('Tổng hóa đơn (USD)')
plt.ylabel('Tiền tip (USD)')
plt.show()
```

```python
# 6. So sánh số tiền mà khách hàng đã chi giữa các ngày trong tuần.
total_tip_by_day = tips.groupby('day')['total_bill'].sum()
plt.figure(figsize=(8,5))
total_tip_by_day.plot(kind='bar', color='skyblue')
plt.title('So sánh số tiền mà khách hàng đã chi giữa các ngày trong tuần')
plt.xlabel('Ngày')
plt.ylabel('Tổng tiền tip (USD)')
plt.show()
```

```python
# Tính tổng hóa đơn theo từng ngày
total_bill_by_day = tips.groupby('day')['total_bill'].sum()

# Vẽ biểu đồ cột
plt.figure(figsize=(8, 5))
total_bill_by_day.plot(kind='bar', color='lightcoral')
plt.title('Tổng chi tiêu theo ngày trong tuần')
plt.xlabel('Ngày')
plt.ylabel('Tổng hoá đơn (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## Với Seaborn

Bạn hãy tìm hiểu và trực quan với thư viện seaborn

```python
import seaborn as sns
tips = pd.read_csv('tips.csv')
```

```python
# 1. Vẽ đồ thị thể hiện tổng tiền tip theo ngày
plt.figure(figsize=(8, 5))
sns.barplot(x='day', y='tip', data=tips, estimator=sum)
plt.title('Tổng tiền tip theo ngày')
plt.xlabel('Ngày')
plt.ylabel('Tổng tiền tip (USD)')
plt.show()
```

```python
# 2. Vẽ đồ thị thể hiện tổng tiền tip theo bữa ăn (time)
plt.figure(figsize=(8, 5))
sns.barplot(x='time', y='tip', data=tips, estimator=sum)
plt.title('Tổng tiền tip theo bữa ăn')
plt.xlabel('Buổi ăn')
```

```python
# 3. Vẽ đồ thị histogram của tổng hoá đơn
plt.figure(figsize=(8,5))
sns.histplot(tips['total_bill'], bins=20, kde=True, color='skyblue')
plt.title('Histogram của tổng hoá đơn')
plt.xlabel('Tổng hoá đơn (USD)')
```

```python
# 4. Vẽ đồ thị boxplot của tổng hoá đơn
plt.figure(figsize=(8,5))
sns.boxplot(x='total_bill', data=tips, color='skyblue')
plt.title('Boxplot của tổng hóa đơn')
plt.xlabel('Tổng hóa đơn (USD)')
```

```python
# 5. Vẽ đồ thị để kiểm tra sự liên quan giữa tổng hoá đơn và tiền tip.
plt.figure(figsize=(8,5))
sns.scatterplot(x='total_bill', y='tip', data=tips, color='blue', alpha=0.5)
plt.title('Sự liên quan giữa tổng hóa đơn và tiền tip')
```

```python
# 6. So sánh số tiền mà khách hàng đã chi giữa các ngày trong tuần.
plt.figure(figsize=(8,5))
sns.barplot(x='day', y='total_bill', data=tips, estimator=sum)
plt.title('So sánh số tiền mà khách hàng đã chi giữa các ngày trong tuần')
```
