## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

````python
import numpy as np
from numpy import random
print(np.__version__)

# Phát sinh ngẫu nhiên mảng arr gồm 100 phần tử số nguyên có giá trị từ 0 đến 1000
arr = np.random.randint(0, 1001, 100)
print("Mảng ngẫu nhiên arr =\n",arr)
````

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

````python
#Câu 1
# 1. Hiển thị 10 phần tử đầu tiên
print("10 phần tử đầu tiên:", arr[:10])

# 2. Hiển thị 5 phần tử cuối cùng
print("5 phần tử cuối cùng:", arr[-5:])

# 3. Hiển thị các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100
result = arr[(arr % 5 == 0) & (arr > 100)]
print("Các phần tử chia hết cho 5 và lớn hơn 100:", result)
````

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

````python
#Câu 2
count21 = 0
result21 = []
for i in arr:
    if i < 100:
        count21+=1
        result21.append(i)

print("Có %d phần tử có giá trị nhỏ hơn 100:" %count21)
print(result21)
        
````

````python
#Câu 2 cách khác:
# Đếm số phần tử có giá trị nhỏ hơn 100
count22 = np.sum(arr < 100)

# Lấy các phần tử có giá trị nhỏ hơn 100
result22 = arr[arr < 100]

print("Có %d phần tử có giá trị nhỏ hơn 100:" % count22)
print(result22)
````

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

````python
#Tính tổng mảng
print(f'- Tổng: {arr.sum()}')

#Tìm giá trị trung bình
print('- Trung bình:', arr.mean())

#Tìm giá trị trung vị
print('- Trung vị:', np.median(arr))

#Tìm Q1, Q2, Q3
print('- Tứ phân vị [Q1, Q2, Q3]:', np.quantile(arr, [0.25, 0.5, 0.75]))

#Tìm phân vị 20 (percentile 20)
print('- Phân vị 20:', np.percentile(arr, 20))
````

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

````python
import numpy as np
# Chọn ngẫu nhiên 10 phần tử không hoàn lại từ mảng arr 5 lần
for i in range(5):
    selected = np.random.choice(arr, size=10, replace=False)
    print(f'Lần chọn ngẫu nhiên {i+1}: {selected}')
````

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

````python
import numpy as np

# Tìm phần tử xuất hiện nhiều nhất
unique, counts = np.unique(arr, return_counts=True)
most_frequent_element = unique[np.argmax(counts)]

print(f'Phần tử xuất hiện nhiều nhất trong mảng là: {most_frequent_element}')
````

### Câu 10: Kết hợp trên cùng một đồ thị

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ hình thể hiện phân phối như sau:

Hướng dẫn: thêm tham số `vert=False` trong hàm vẽ `boxplot()`

````python
# Your code here
````

````python
#Câu 10: Kết hợp trên cùng một đồ thị
import numpy as np
import matplotlib.pyplot as plt

# Phát sinh 10000 số ngẫu nhiên thuộc phân phối chuẩn
data = np.random.normal(size=10000)

# Tạo một figure với 2 subplot
fig = plt.figure()
fig.suptitle('Đồ thị hàm sin(x) và cos(x)')

#với 2 subplot
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)


#ax1 Vẽ histogram của các số ngẫu nhiên #Vẽ histogram sử dụng hàm hist()
ax1.hist(data, bins=100,
         color='purple',
         edgecolor='white',
        density=True) #bins: chia thành 100 cột
#ax1 thêm nhãn và lable ax1 histogram
ax1.set_title('Histogram')
ax1.set_ylabel('Mật độ')

#ax2 Vẽ boxplot
ax2.boxplot(x=data, vert=False)
ax2.set_title('Boxplot')

# Hiển thị đồ thị
plt.tight_layout() #plt.tight_layout() điều chỉnh để các subplot không bị chồng lên nhau.
plt.show()
````

### 1. Vẽ đồ thị thể hiện tổng tiền tip theo ngày

````python
import matplotlib.pyplot as plt 
import seaborn as sns

# Tải dữ liệu tips
df = sns.load_dataset("tips")

# Vẽ đồ thị thể hiện tổng tiền tip theo ngày
sns.barplot(x="day", y="tip", data=df, estimator=sum, errorbar=None)

# Định dạng và hiển thị đồ thị
plt.title('Tổng tiền tip theo ngày')
plt.xlabel('Ngày')
plt.ylabel('Tổng tiền tip')
plt.show()

# x="day": Trục x biểu diễn các ngày trong tuần.
# y="tip": Trục y biểu diễn giá trị tiền tip.
# data=df: Dữ liệu được sử dụng là DataFrame df.
# estimator=sum: Sử dụng hàm sum để tính tổng tiền tip cho mỗi ngày.
# ci=None: Loại bỏ khoảng tin cậy (confidence interval) cho biểu đồ thanh.
# errorbar=None: nếu không hỗ trợ ci=None thì dùng errorbar để thay thế
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Tải dữ liệu tips
df = sns.load_dataset("tips")

# Tạo subplot để hiển thị hai đồ thị histogram cạnh nhau
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Vẽ histogram cho thời gian dùng bữa "Lunch"
sns.histplot(df[df["time"] == "Lunch"]["total_bill"], kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Tổng tiền hóa đơn cho bữa trưa (Lunch)')
axes[0].set_xlabel('Tổng tiền hóa đơn')
axes[0].set_ylabel('Tần suất')

# Vẽ histogram cho thời gian dùng bữa "Dinner"
sns.histplot(df[df["time"] == "Dinner"]["total_bill"], kde=True, ax=axes[1], color='salmon')
axes[1].set_title('Tổng tiền hóa đơn cho bữa tối (Dinner)')
axes[1].set_xlabel('Tổng tiền hóa đơn')

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
````

### 10. Vẽ hai đồ thị boxplot cạnh nhau để so sánh tổng hóa đơn giữa hai thời gian dùng bữa `Lunch` và `Dinner`

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Tải dữ liệu tips
df = sns.load_dataset("tips")

# Tạo subplot để hiển thị hai đồ thị boxplot cạnh nhau
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Vẽ boxplot cho thời gian dùng bữa "Lunch"
sns.boxplot(x="total_bill", data=df[df["time"] == "Lunch"], ax=axes[0], color='skyblue')
axes[0].set_title('Tổng tiền hóa đơn cho bữa trưa (Lunch)')
axes[0].set_xlabel('Tổng tiền hóa đơn')

# Vẽ boxplot cho thời gian dùng bữa "Dinner"
sns.boxplot(x="total_bill", data=df[df["time"] == "Dinner"], ax=axes[1], color='salmon')
axes[1].set_title('Tổng tiền hóa đơn cho bữa tối (Dinner)')
axes[1].set_xlabel('Tổng tiền hóa đơn')

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
````
