## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

```python
import numpy as np
from numpy import random
arr = np.random.randint(0, 1001, 100)
print(arr)
```

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
print("10 phan tu dau tien: ",arr[:10])
print("5 phan tu cuoi cung: ", arr[5:])
print(arr[(arr%5==0) & (arr>100)])
```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
arr_1 = arr[arr<100]
print(len(arr_1))
print(arr_1)
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
print("tong mang: ", sum(arr))
print("trung binh: ", np.mean(arr))
print("trung vi: ", np.median(arr))
print("phan vi q1: ", np.percentile(arr, 25))
print("phan vi q2: ", np.percentile(arr, 50))
print("phan vi q3: ", np.percentile(arr, 75))
print("phan vi 20: ", np.percentile(arr, 20))
```

### 4. Thực hiện việc chọn ngẫu nhiên (không thay thế) 10 phần tử từ mảng trên 5 lần. Hiển thị kết quả lựa chọn mỗi lần

```python
for i in range(5):
    print(np.random.choice(arr, size=10, replace=False))
```

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

```python
val, count = np.unique(arr, return_counts=True)
ans = val[np.argmax(count)]
print("phan tu xuat hien nhieu nhat: ", ans)
print("so lan xuat hien: ", count.max())
```

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE
x = np.linspace(0, 6.28, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label = 'sin(x)')
plt.plot(x, y2, linestyle = '--', label = 'cos(x)')
plt.legend()
plt.title('do thi ham sin(x) cos(x)')
plt.show()
```
